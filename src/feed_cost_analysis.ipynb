```python
import pandas as pd
import pymysql  # 또는 psycopg2 for PostgreSQL

# RDS 접속 정보 (예시)
host = 'database-1.cba44miw26by.ap-northeast-2.rds.amazonaws.com'
port = 3306
user = 'psy'
password = '0000'
database = 'pig'

# 연결
conn = pymysql.connect(
    host=host,
    port=port,
    user=user,
    password=password,
    database=database,
    charset='utf8mb4'
)

# SQL 쿼리 실행
query = "SELECT * FROM feed"
df = pd.read_sql(query, conn)

# 연결 종료
conn.close()
```

```python
yearly = df.groupby('year').mean(numeric_only=True).reset_index()
```

```python
import matplotlib.pyplot as plt
import matplotlib_koreanize
import seaborn as sns

plt.figure(figsize=(10, 5))
sns.lineplot(data=yearly, x='year', y='사료비', marker='o')
plt.title('연도별 사료비 추이')
plt.grid(True)
plt.show()
```

```python
sns.scatterplot(data=yearly, x='corn_price', y='사료비')
plt.title('옥수수 가격 vs 사료비')
plt.show()
```

```python
sns.scatterplot(data=yearly, x='exchange_rate', y='krw_based_price')
plt.title('환율 vs 원화 기준 옥수수 가격')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 연도별 평균 데이터로 정리
yearly = df.groupby('year').mean(numeric_only=True).reset_index()

plt.figure(figsize=(12, 6))
plt.plot(yearly['year'], yearly['corn_price'], label='옥수수 가격 (USD)', marker='o')
plt.plot(yearly['year'], yearly['exchange_rate'], label='환율', marker='o')
plt.plot(yearly['year'], yearly['krw_based_price'], label='원화 옥수수 가격', marker='o')
plt.plot(yearly['year'], yearly['사료비'], label='사료비', marker='o')
plt.title("연도별 주요 변수 비교")
plt.xlabel("연도")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(data=yearly, x='year', y='Multiplier_of_cost', marker='o')
plt.title("연도별 사료 단가 배수 변화 (Multiplier of Cost)")
plt.ylabel("배수")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(data=yearly, x='year', y='사료비', color='skyblue')
plt.title("연도별 평균 사료비")
plt.ylabel("사료비 (KRW)")
plt.xticks(rotation=45)
plt.grid(True, axis='y')
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.regplot(data=yearly, x='corn_price', y='사료비')
plt.title("옥수수 가격과 사료비 간의 관계")
plt.xlabel("옥수수 가격 (USD)")
plt.ylabel("사료비 (KRW)")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.scatterplot(data=yearly, x='exchange_rate', y='Multiplier_of_cost')
plt.title("환율과 사료 단가 배수의 관계")
plt.xlabel("환율")
plt.ylabel("Multiplier of Cost")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
corr = yearly.drop(columns='year').corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("연도별 주요 변수 간 상관관계")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(data=yearly, x='year', y='corn_1kg', label='옥수수 1kg 가격', marker='o')
sns.lineplot(data=yearly, x='year', y='사료비', label='사료비', marker='o')
plt.title("사료비 vs 옥수수 1kg 가격")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

ticker = yf.Ticker("ZC=F")

data = ticker.history(period="1y")

plt.figure(figsize=(14, 6))
plt.plot(data.index, data['Close'], label='Corn Futures (ZC=F)', color='red')

plt.title("Volatility in Corn Futures Prices", fontsize=15)
plt.xlabel("date")
plt.ylabel("price (USD)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
